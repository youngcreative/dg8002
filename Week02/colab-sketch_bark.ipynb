```python
#@title **Bark Demo 🐕**
#@markdown # Bark demo

#@markdown Bark is a transformer-based text-to-audio model created by [Suno](https://suno.ai). Bark can generate highly realistic, multilingual speech as well as other audio — including music, background noise and simple sound effects. The model can also produce nonverbal communications like laughing, sighing and crying.

#@markdown 1. Select the voice.

voicePreference = 'Random' #@param ["Random", "Man", "Woman"]

if voicePreference == 'Random':
    voicePreference = ''
elif voicePreference == 'Man':
    voicePreference = 'MAN: '
else:
    voicePreference = 'FEMALE: '


#@markdown 2. Choose whether this is a song lyric

songLyrics = False             #@param {type:"boolean"}

#@markdown 3. Enter the voice prompt below. Here are some tips:
#@markdown  - EMPHASIZE text by using UPPERCASE
#@markdown  - The following non-text interjections are possible: [laughter], [laughs], [sighs], [music], [gasps], [clears throat]

barkTextPrompt = "I may be [laughter] barking up the wrong tree again."  #@param {type:"string"}

#@markdown 4. Optional: Download the AI-generated voice output as a file.

downloadAsFile = False             #@param {type:"boolean"}


# generate audio from text
if songLyrics == True:
  text_prompt = '♪ ' + voicePreference + barkTextPrompt + ' ♪'
else:
  text_prompt = voicePreference + barkTextPrompt

print('Generating: ' + text_prompt) # for debugging

audio_array = generate_audio(text_prompt)


# save audio to disk
if downloadAsFile == True:
  write_wav("bark_generation.wav", SAMPLE_RATE, audio_array)
  files.download("bark_generation.wav")

# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE)

```
